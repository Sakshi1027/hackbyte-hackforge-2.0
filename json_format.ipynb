```python
import json

FILE = "memory.json"

# Load JSON
with open(FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

# Ensure it's a list
if not isinstance(data, list):
    raise ValueError("JSON root is not a list!")

# Delete items index 0 to 10
del data[0:11]   # deletes elements 0,1,2,...,10

# Save back
with open(FILE, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("Deleted entries 0–10 successfully!")
```
